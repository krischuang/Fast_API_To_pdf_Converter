```python
import os
import img2pdf

def convert_pngs_to_pdf(input_dir, output_pdf_path):
    """
    Converts all PNG images in a directory to a single PDF file.

    Args:
        input_dir (str): The path to the directory containing PNG images.
        output_pdf_path (str): The path and filename for the output PDF.
    """
    png_files = []
    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".png"):
            png_files.append(os.path.join(input_dir, filename))

    # Sort the files to ensure consistent page order in the PDF
    png_files.sort()

    if not png_files:
        print(f"No PNG images found in '{input_dir}'")
        return

    with open(output_pdf_path, "wb") as f:
        f.write(img2pdf.convert(png_files))
    print(f"Successfully converted PNGs in '{input_dir}' to '{output_pdf_path}'")

# Example usage:
input_directory = "."  # Replace with your directory
output_pdf_file = "output.pdf"
convert_pngs_to_pdf(input_directory, output_pdf_file)
```
